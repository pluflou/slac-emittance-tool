# Get beamsizes (in pixels)

```python
# xrms, yrms, xrms_error, yrms_error
beam_image.get_sizes(show_plots=True)
```

```python
def getbeamsizes():
    beam_image = Image(otr_im, n_col, n_row, bg_image = None)
    beam_image.reshape_im()
    beam_image.subtract_bg()
    beam_image.get_im_projection()
    return beam_image.get_sizes(show_plots=False)
getbeamsizes()
```
